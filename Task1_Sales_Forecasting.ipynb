### Data Preprocessing

```python
# select required columns
data = df[['order_date', 'total']]

# convert date
data['order_date'] = pd.to_datetime(data['order_date'])

# group by date
data = data.groupby('order_date').sum().reset_index()

# sort
data = data.sort_values('order_date')

data.head()
```

### Feature Engineering

```python
data['date_ordinal'] = data['order_date'].map(lambda x: x.toordinal())
```

### Data Visualization

```python
plt.figure(figsize=(10,5))
plt.plot(data['order_date'], data['total'])
plt.title("Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.show()
```

### Model Building

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X = data[['date_ordinal']]
y = data['total']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

model = LinearRegression()
model.fit(X_train, y_train)
```

### Model Prediction

```python
predictions = model.predict(X_test)
```

### Model Evaluation

```python
from sklearn.metrics import r2_score

print("R2 Score:", r2_score(y_test, predictions))
```

### Actual vs Predicted

```python
plt.figure(figsize=(10,5))
plt.plot(y_test.values, label="Actual")
plt.plot(predictions, label="Predicted")
plt.legend()
plt.show()
```

### Improved Model - Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=100)

model.fit(X_train, y_train)

predictions = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score

print("R2 Score:", r2_score(y_test, predictions))
```

```python
model = RandomForestRegressor(n_estimators=100, max_depth=8, min_samples_split=5, random_state=42)
model.fit(X_train, y_train)
```

```python
# Make predictions with the tuned model
new_predictions = model.predict(X_test)

# Calculate and print the new R2 Score
new_r2 = r2_score(y_test, new_predictions)
print(f"New R2 Score: {new_r2}")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

# 1. Date features ni correct column name ('order_date') tho extract cheddam
df['order_date'] = pd.to_datetime(df['order_date'])
df['month'] = df['order_date'].dt.month
df['day_of_week'] = df['order_date'].dt.dayofweek
df['is_weekend'] = df['day_of_week'].apply(lambda x: 1 if x >= 5 else 0)

# 2. Features (X) and Target (y) update cheddam
# Target column 'Sales' kadhu, mee list lo 'total' ani undhi
X = df[['month', 'day_of_week', 'is_weekend', 'qty_ordered', 'price']] 
y = df['total'] 

# 3. Train-Test Split (Sales data ki shuffle=False chala important)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# 4. Model Training with constraints (Overfitting thagginchadaniki)
model = RandomForestRegressor(n_estimators=100, max_depth=10, min_samples_split=5, random_state=42)
model.fit(X_train, y_train)

# 5. Final Score check
r2_score = model.score(X_test, y_test)
print(f"Final R2 Score: {r2_score}")

# Columns ela unnao chudali ante:
print("\nUsed Features:", X.columns.tolist())
```
